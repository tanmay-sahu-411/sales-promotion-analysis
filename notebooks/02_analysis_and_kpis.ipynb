# 📊 Sales Promotion Analysis

## Objective
Analyze the impact of promotional discounts on sales performance and understand product, region, and discount effectiveness.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/promotion_data.csv")
df.head()
```

```python
df.info()
```

```python
df["Revenue"] = df["Units_Sold"] * df["Promo_Price"]
df["Discount_%"] = ((df["Original_Price"] - df["Promo_Price"]) / df["Original_Price"]) * 100

df.head()
```

```python
total_revenue = df["Revenue"].sum()
total_units = df["Units_Sold"].sum()
avg_discount = df["Discount_%"].mean()

print("Total Revenue:", total_revenue)
print("Total Units Sold:", total_units)
print("Average Discount:", round(avg_discount, 2))
```

```python
df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("City")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Product")["Units_Sold"].sum().sort_values(ascending=False)
```

```python
df.groupby("Product")[["Discount_%", "Units_Sold"]].mean()
```

```python
df.groupby("Date")["Revenue"].sum()
```

```python
df.to_csv("../data/promotion_data_with_kpis.csv", index=False)
```

# 📊 Key Insights

## 1. Product Performance
Headphones generated the highest revenue, indicating strong demand during promotions. This suggests that certain products respond better to discounts.

## 2. Regional Performance
Metro cities like Bangalore and Delhi contributed more to total revenue, showing that promotions are more effective in high-demand urban markets.

## 3. Discount Impact
Higher discounts increased units sold, but the increase was not always proportional. This indicates that very high discounts may not always be necessary.

## 4. Sales Trend
Sales varied across different days, with certain days showing peak performance. This suggests timing plays an important role in promotions.

## 5. Overall Insight
Promotions successfully increased demand, but optimizing discount levels and focusing on high-performing products and regions can improve business outcomes.

## 📌 Conclusion
Promotions increase demand, but their effectiveness depends on product type, region, and discount strategy. Businesses should focus on optimizing discounts rather than maximizing them.
